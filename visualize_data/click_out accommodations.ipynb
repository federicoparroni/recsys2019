## Average accomodations price

```python
clickouts_df = train_df
for idx, row in tqdm(clickouts_df[clickouts_df.impressions.notnull()].iterrows()):
    impressions = [int(x) for x in row.impressions.split('|')]
    print(row.prices.split('|'))
    prices = [int(x) for x in row.prices.split('|')]
    for i in range(len(impressions)):
        dict_accmds[impressions[i]] += prices[i]
```
