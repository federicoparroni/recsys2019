```python
import sys 
import os 
os.chdir("/Users/Albo/Documents/GitHub/keyblade95/recsys2019")
os.getcwd()
from tqdm.auto import tqdm as tqdm
import pandas as pd 
import numpy as np 
import data
tqdm.pandas()
from preprocess_utils.last_clickout_indices import find
```

```python
full = data.full_df()
```

```python
last_clickout_indices = find(full)
df_dropped_last_clickouts = full.drop(last_clickout_indices)
cf = df_dropped_last_clickouts[df_dropped_last_clickouts.action_type == 'clickout item']['current_filters'].dropna()
cf = cf.progress_apply(lambda x: x.split('|'))
unique_filters = np.unique([sub for f in cf for sub in f])
d = df_dropped_last_clickouts.loc[cf.index]
references = d.reference
dict_filter_pop_ref = dict(zip(unique_filters,np.zeros(len(unique_filters))))
for i in dict_filter_pop_ref.keys():
    dict_filter_pop_ref[i] = dict(zip(references.unique(), np.zeros(len(references.unique()))))
for index, row in tqdm(d.iterrows(),total=d.shape[0]):
    for f in cf.loc[index]:
        dict_filter_pop_ref[f][row.reference] +=1
final = pd.DataFrame.from_dict(dict_filter_pop_ref).astype(int)
```
